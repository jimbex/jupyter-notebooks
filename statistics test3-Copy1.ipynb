```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv('FuelConsumptionCo2.csv')

x = df['FUELCONSUMPTION_HWY']
y = df['CO2EMISSIONS']

slope, intercept, r, p, std_err = stats.linregress(x, y)

def mine(x):
    return slope * x + intercept
mymodel = list(map(mine, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
train_x = x[:854]
train_y = y[:854]

test_x = x[854:]
test_y = y[854:]

```

```python
slope, intercept, r, p, std_err = stats.linregress(train_x, train_y)
def mine(x):
    return slope * x + intercept
mymodel = list(map(mine, train_x))

plt.scatter(train_x, train_y)
plt.plot(train_x, mymodel)
plt.show()
r
```

```python
slope, intercept, r, p, std_err = stats.linregress(test_x, test_y)
def mine(x):
    return slope * x + intercept
mymodel = list(map(mine, test_x))

plt.scatter(test_x, test_y)
plt.plot(test_x, mymodel)
plt.show()
r
```
